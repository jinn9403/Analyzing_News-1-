''.join(리스트)
-''.join(리스트)를 이용하면 매개변수로 들어온 ['a', 'b', 'c'] 이런 식의 리스트를 'abc'의 문자열로 합쳐서 반환해주는 함수

'구분자'.join(리스트)
-'구분자'.join(리스트)를 이용하면 리스트의 값과 값 사이에 -'구분자'에 들어온 구분자를 넣어서 하나의 문자열로 합친다
-'_'.join(['a', 'b', 'c']) 라 하면 "a_b_c" 와 같은 형태로 문자열을 만들어서 반환한다

re 라이브러리
-정규표현식(regular expressions)의 약자로, 복잡한 문자열을 처리할 때 사용하는 기법으로, 파이썬 등 다양한 문언어에서 사용
-특정문자열 추출, 변환 등에 사용
-stackoverflow에서 자세한 사용법 확인

```python
import re
```

```python
words = re.findall(r'\w+', text)
#\w단어를 추출하는 정규표현식
print(words)
```

```python
#전화번호 추출

contact = ''' 김미키 21 010-1231-1231 miky@gmail.com
              이아름 29 010-2223-8211 ekshd@gmail.com'''
```

```python
regex = r'0\d{1,2}[-]?\d{3,4}[-]?\d{3,4}'
#0 : 0으로 시작, d : digit, {1,2} : 1~2자리 수 ...

phone = re.findall(regex, contact)
print('\n'.join(phone))
```

```python
#문자열 변환(마스킹)
pat = re.compile(r'0\d{1,2}[-]?\d{3,4}[-]?\d{3,4}')
print(pat.sub("***-****-****", contact))
```

단어 갯수 구하기

-collections.Counter

-리스트나 문자열과 같은 자료형의 요소 중 값이 같은 요소가 몇 개인지를 확인할 때 사용

```python
import collections
import textwrap
import re
```

```python
#textwrap.fill()
filled_text = textwrap.fill(text, width = 40)
print(filled_text)
#위에 두 과정을 합한 것
```

```python
words = re.findall(r'\w+', text)
#\w단어를 추출하는 정규표현식
print(words)
```

```python
#counter()
#빈도수 산출
counter = collections.Counter(words)
print(counter)
#딕션너리 형태로 덧셈이 가능하다.
```

```python
#덧셈
a = collections.Counter(['a', 'b', 'b', 'c', 'd', 'a'])
b = collections.Counter(['a', 'f', 'f', 'b', 'd', 'a'])
```

```python
print(a)
print(b)
```

```python
#사칙연산
print(a-b)
print(a & b)
print(a | b)
```

```python
#most_common()
#상위 빈도수 단어 추출
print(counter.most_common(10))
#상위 n개 갯수 지정 가능
```
